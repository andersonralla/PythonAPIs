```python
r2 = requests.get('https://api.github.com/versions')
print(r2.status_code)
```

```python
r2.json()
```

```python
# especificando a versão da API
version = {'X-GitHub-Api-Version': '2022-11-28'}
r3 = requests.get('https://api.github.com/events', headers=version)
```

```python
print(r3.status_code)
r3.json()
```
